```python
"""
Strongly focused on classification (or classification-heavy applications)

Rajpurkar et al. (2017) – CheXNet: Radiologist-level pneumonia detection on chest X-rays with deep learning
This is a pure classification paper. It trains a deep convolutional neural network (121-layer DenseNet) to classify chest X-rays for the presence/absence of pneumonia (and extends to 14 binary/multi-label disease classification tasks). It explicitly compares model performance against radiologists using classification metrics (F1-score, etc.). This is one of the clearest examples of an AI classification problem in the list.
Strøm et al. (2018) – Machine Learning Performance Metrics and Diagnostic Context in Radiology
This paper discusses performance metrics for machine learning in a diagnostic (classification) context in radiology. It is directly paired with the CheXNet paper in the assignment and focuses on how metrics should be chosen and interpreted for classification tasks in medical imaging.
Sokolova et al. (2006) – Beyond accuracy, F-score and ROC: a family of discriminant measures for performance evaluation
This is one of the most directly relevant papers for classification. It reviews and proposes performance measures specifically for classification problems (including medical diagnosis), arguing that accuracy, F-score, and ROC are not always sufficient. It introduces additional discriminant measures (Youden’s index, likelihood ratios, discriminant power, etc.) that are especially useful when classes are imbalanced or when different error types have different costs.

Papers that cover classification as a major topic (alongside regression or general ML)

LeCun, Bengio & Hinton (2015) – Deep learning (Nature review)
This seminal review discusses deep learning in general, but it heavily features classification successes (especially image classification with convolutional nets, speech recognition, etc.). It explains how higher layers in deep networks are particularly good at building representations for discrimination/classification tasks.
Ng, Andrew (2016) – What artificial intelligence can and can’t do right now (Harvard Business Review)
Ng explicitly highlights that most practical AI today comes from supervised learning, and within that, especially classification and regression. He uses many classification examples (image recognition, spam detection, medical diagnosis, etc.) to illustrate what current AI excels at.

Others that are more general or only tangentially related

The papers on data science process (Goldstein, Zhu & Xiong, Tayo on problem framing), statistics, deep learning fundamentals (Leshno on activation functions, LeCun on backprop), decision trees/random forests, clustering, anomaly detection, attention/transformers, etc., are useful background but do not focus specifically on classification over regression or on classification-specific techniques.

Summary – Recommended reading order for your question
If you want to focus on AI classification problems and techniques:

CheXNet (Rajpurkar et al.) – best concrete example of a real-world deep learning classification system.
Sokolova et al. – best for understanding which metrics and considerations matter specifically for classification.
Strøm et al. – ties the two together in a medical diagnostic (classification) context.
LeCun et al. (Deep Learning) and Ng (2016) – excellent broader context on why classification is so central to current AI success.
"""
```
